### Sum of Numbers from 1 to 10

The code in the next cell calculates the sum of numbers from 1 to 10 using the `sum` function and the `range` function in Python. Let's display the result.


```python
# Calculate the sum of numbers from 1 to 10
total_sum = sum(range(1, 11))
total_sum
```

```python
import numpy as np

# Numpy Examples

# Creating a Numpy array
arr = np.array([1, 2, 3, 4, 5])
print("Numpy Array:", arr)

# Performing operations on the array
mean_value = np.mean(arr)
print("Mean:", mean_value)
```

```python
# Matplotlib Examples
import matplotlib.pyplot as plt


# Creating a simple plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sinusoidal Plot')
plt.show()
```

```python
# MNE-Python Examples
import mne
import os 

path = os.path.join(mne.datasets.sample.data_path(), 'MEG', 'sample', 'sample_audvis_raw.fif')

# Loading example MEG data
sample_data_path = mne.datasets.sample.data_path()
raw = mne.io.read_raw_fif(path, preload=True)

# Plotting the raw data
raw.plot()

# Selecting and plotting a specific range of time
start, stop = raw.time_as_index([10, 20])  # Selecting data between 10s and 20s
data, times = raw[:, start:stop]
plt.plot(times, data.T)
plt.xlabel('Time (s)')
plt.ylabel('MEG data (T)')
plt.title('MEG Data between 10s and 20s')
plt.show()
```
