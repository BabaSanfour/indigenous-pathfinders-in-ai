### Sum of Numbers from 1 to 10

The code in the next cell calculates the sum of numbers from 1 to 10 using the `sum` function and the `range` function in Python. Let's display the result.


```python
# Calculate the sum of numbers from 1 to 10
total_sum = sum(range(1, 11))
total_sum
```

### Now We have the examples from module 2.3

```python
import numpy as np

# Numpy Examples

# Creating a Numpy array
arr = np.array([1, 2, 3, 4, 5])
print("Numpy Array:", arr)

# Performing operations on the array
mean_value = np.mean(arr)
print("Mean:", mean_value)
```

```python
# Pandas Examples
import pandas as pd
# Creating a Pandas DataFrame
data = {'Name': ['John', 'Alice', 'Bob'], 'Age': [25, 30, 22]}
df = pd.DataFrame(data)
print("Pandas DataFrame:")
print(df)

# Accessing columns in Pandas DataFrame
ages = df['Age']
print("Ages Column:")
print(ages)
```

```python
# Matplotlib Examples
import matplotlib.pyplot as plt


# Creating a simple plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sinusoidal Plot')
plt.show()
```
